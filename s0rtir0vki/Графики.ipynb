```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as so
import csv


x_data, y_data = [], []
with open("bubble_sort.csv", encoding='utf-8') as r_file:
    
    file = csv.reader(r_file, delimiter = ";")
    count = 0
    for row in file:
        if count == 0:
            pass
        else:
            x_data.append(row[0])
            y_data.append(row[1])
        count += 1

                      
for i in range(0, len(x_data)):
    x_data[i] = float(x_data[i])
for i in range(0, len(y_data)):
    y_data[i] = float(y_data[i])

def f(x, a):
    y = a * x * x
    return y


G = so.curve_fit(f, x_data, y_data)
a1 = G[0][0]


y_new, x_new = [], []
for i in range(0, int(x_data[len(x_data)-1]), 20):
    x_new.append(i)
    y_new.append(f(i, a1))

plt.figure(figsize=(18, 10))
plt.plot(x_data, y_data, 'ro', label=r'experimental', color='orange')
plt.plot(x_new, y_new, label=r'theoretical', color='black', linewidth='3')
plt.xlabel(r'$N$', fontsize=14)
plt.ylabel(r'$Time,\  s$', fontsize=14)
plt.title(r'$bubble\  sort$', fontsize=16)
plt.legend(loc='best', fontsize=12)
plt.grid(True)

a1_string = str(round(a1*1000000, 1))
plt.text(0, 950, ' : t = ' + a1_string + 'e-06 * N^2', fontsize = 20)
                      
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as so
import csv


x_data, y_data = [], []
with open("std_sort.csv", encoding='utf-8') as r_file:
    
    file = csv.reader(r_file, delimiter = ";")
    count = 0
    for row in file:
        if count == 0:
            pass
        else:
            x_data.append(row[0])
            y_data.append(row[1])
        count += 1

                      
for i in range(0, len(x_data)):
    x_data[i] = float(x_data[i])
for i in range(0, len(y_data)):
    y_data[i] = float(y_data[i])

def f2(x, a):
    y = a * x * np.log(x)
    return y


G = so.curve_fit(f2, x_data, y_data)
a6 = G[0][0]


y_new, x_new = [], []
for i in range(0, int(x_data[len(x_data)-1]), 20):
    x_new.append(i)
    y_new.append(f2(i, a6))

plt.figure(figsize=(18, 10))
plt.plot(x_data, y_data, 'ro', label=r'experimental', color='orange')
plt.plot(x_new, y_new, label=r'theoretical', color='black', linewidth='3')
plt.xlabel(r'$N$', fontsize=14)
plt.ylabel(r'$Time,\  s$', fontsize=14)
plt.title(r'$std\  sort$', fontsize=16)
plt.legend(loc='best', fontsize=12)
plt.grid(True)

a6_string = str(round(a6*100000, 1))
plt.text(0, 260, ' : t = ' + a6_string + 'e-05 * NlnN', fontsize = 20)
                      
plt.show()
```
